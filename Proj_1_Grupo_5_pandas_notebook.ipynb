## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
# Verifica quais colunas têm dados faltantes e sua porcentagem
# Existem dados faltantes nas colunas de Ativo 01 a 06 e de Notas
df.isnull().sum() * 100 / len(df)
```

```python
# Trata os dados faltantes das colunas de Ativo, inserindo o número 0 nesses casos
# Os valores de Ativo representam o quanto o investidor possui em cada Ativo e, quando o valor é NaN, entende-se que não há investimento
# Não podemos excluir as linhas porque a quantidade é relevante
for coluna in df.iloc[:, 10:17]:
  df[coluna].fillna(value=0, inplace=True)
```

```python
# Trata os dados faltantes das colunas de Notas, inserindo a moda
# Entende-se que a moda, por ser o valor mais frequente, seria a nota mais provável a ser dada pelos investidores que não avaliaram a empresa
# Outra decisão possível seria excluir as linhas com valores faltantes porque compõem apenas 12% do DataFrame, no máximo (considerando que não estão
# na mesma linha, ou seja, que o mesmo investidor não avaliou a empresa)
df['Nota1'].fillna(df['Nota1'].mode()[0], inplace=True)
df['Nota2'].fillna(df['Nota2'].mode()[0], inplace=True)

# Verifica se os dados foram tratados corretamente
df.isnull().sum() * 100 / len(df)
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50


```python
# Cria coluna "Total" somando o valor de todos os ativos por investidor
df['Total'] = df.iloc[:, 10:17].sum(axis=1)

# Cria coluna "Media" com a média das notas 1 e 2
df['Media'] = (df['Nota1'] + df['Nota2']) / 2

# Cria coluna "Dominio" com apenas o domínio do email (dica: separar por @)
# Encontra o @ como separador, cria uma lista com os domínios e insere no DataFrame
lista_dominio = []
for email in df['Email']:
  dominio = email[email.find('@')+1:]
  lista_dominio.append(dominio)
df['Dominio'] = lista_dominio

# Muda o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
df = df.rename(columns=({'Salario': 'Renda'}))
df.Renda = df.Renda.astype('str')
df['Renda'] = df['Renda'].str.replace('$', '')
df.Renda = df.Renda.astype('float')

# Transforma os dados da coluna Media em inteiro
df.Media = df.Media.astype('int')

# Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, 
#e detrator se for abaixo de 50
lista_nps = []
for i in df['Media']:
  if i >= 80:
    i = 'promotor'
  elif i < 80 and i >= 50:
    i = 'neutro'
  else:
    i = 'detrator'
  lista_nps.append(i)
df['NPS'] = lista_nps
```

```python
# Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
df.Sexo = df.Sexo.astype('str')
lista_genero = list(map(lambda x: x.replace('Fem', 'Feminino').replace('Masc', 'Masculino'), df['Sexo']))
df['Sexo'] = lista_genero
```

- Mostre a média e desvio padrão de cada ativo

```python
# Quantos investidores são promotores? Lembrando: promotor é quem deu nota maior igual a 80
promotores = len(df[(df['NPS'] == 'promotor')])
print(f'{promotores} investidores são promotores, sendo {round((promotores/len(df)*100), 2)}% da população')
print('-' * 32)

# Quantos investidores são do sexo feminino?
feminino = len(df[(df['Sexo'] == 'Feminino')])
print(f'{feminino} investidores do sexo feminino, sendo {round((feminino/len(df)*100), 2)}% da população')

masculino = len(df[(df['Sexo'] == 'Masculino')])
print(f'{masculino} investidores do sexo masculino, sendo {round((masculino/len(df)*100), 2)}% da população')
print('-' * 32)

# Quantos investidores investiram mais de 16000 no Ativo04?
maior_16mil = len(df[(df['Ativo04'] > 16000)])
print(f'{maior_16mil} investidores investiram mais de 16000 no Ativo04, sendo {round((maior_16mil/len(df)*100), 2)}% da população')
print('-' * 32)

# Quantos investidores investiram mais de 16000 no Ativo04 e menos de 5000 no Ativo05?
entre_16mil_5mil = len(df[(df['Ativo04'] > 16000) & (df['Ativo05'] < 5000)])
print(f'{entre_16mil_5mil} investidores investiram mais de 16000 no Ativo04 e menos de 5000 no Ativo05, sendo {round((entre_16mil_5mil/len(df)*100), 2)}% da população')
print('-' * 32)

# Quantos investidores do sexo feminino tem mais de 25 anos e investiu menos de 10000? (bonus: por que esse dado seria importante 
# para a estratégia da nossa empresa?
mulheres_menos10mil = len(df[(df['Sexo'] == 'Feminino') & (df['Idade'] > 25) & (df['Total'] < 10000)])
print(f'{mulheres_menos10mil} investidoras maiores de 25 anos investiram menos de 10000, sendo {round((mulheres_menos10mil/len(df)*100), 2)}% da população')
# Considerando que perfazem somente 1,14% da população, esse dado não é importante para a estratégia da empresa
```

```python
# Quantos investidores tem em cada estado (dica: use groupby)
# Utilizamos o método size porque tratamos os valores nulos anteriormente
qntde_investidores_estado = df.groupby('Estado').size()

# Qual é a soma do total investido por cada estado?
valor_investidores_estado = df.groupby('Estado')[['Total']].sum()

# Junta as informações acima, acrescenta a porcentagem, e ordena as colunas para melhor visualização
investimento_estado = pd.concat([qntde_investidores_estado, valor_investidores_estado], axis=1)

investimento_estado['% Qtde'] = (qntde_investidores_estado / sum(qntde_investidores_estado)) * 100
investimento_estado['% Total'] = (valor_investidores_estado / valor_investidores_estado.sum()) * 100

investimento_estado = investimento_estado.rename(columns={
    0: 'Qntde investidores',
    'Total': 'Total investimento'
})

investimento_estado = investimento_estado[['Qntde investidores','% Qtde', 'Total investimento', '% Total']]
display(investimento_estado)
```

```python
# Como os investidores se distribuem em promotores, detratores e neutros?
distribuicao_nps_size = df.groupby('NPS').size()
# DataFrame referente à porcentagem dos valores.
distribuicao_nps_size_por = pd.DataFrame(
    (round(distribuicao_nps_size/len(df), 2))
    .apply(lambda x: f"{x*100:.2f}%")
)
# Criação do Dataframe com as quantidades
distribuicao_nps_size = pd.DataFrame(distribuicao_nps_size)
# Criando a coluna porcentagem e inserindo os valores do DataFrame Porcentagem
distribuicao_nps_size['Porcentagem'] = distribuicao_nps_size_por
distribuicao_nps_size = distribuicao_nps_size.rename(columns=({0: 'Qtde'}))
# Printando
print('Quantidade de investidores por NPS:')
display(distribuicao_nps_size)
print('-' * 32)
```

```python
# Qual é a média do total investido por promotores, detratores e neutros?
# Groupby referente à media do total
distribuicao_nps_mean = (df
    .groupby(['NPS'])
    .Total
    .mean()
)
# Criação do DataFrame referente às porcentagens por "classe" de investidor
distribuicao_nps_mean_por = pd.DataFrame(
    (distribuicao_nps_mean/distribuicao_nps_mean.sum())
    .apply(lambda x: f"{x*100:.2f}%")
)
# Criação do DataFrame das Médias
distribuicao_nps_mean = pd.DataFrame(round(distribuicao_nps_mean, 3))
# Inclusão do DataFrame na coluna de porcentagem
distribuicao_nps_mean['Porcentagem'] = distribuicao_nps_mean_por
# Printando
print('Média do total investido por NPS:')
display(distribuicao_nps_mean)

print('-' * 32)
```
